### `Q-1:` Using `groupby` make groups using the `"Pclass"` column and find out the average age and total number of missing values in the `"Age"` column for every group.

````python
pclass = titanic_df.groupby('Pclass')
print(pclass['Age'].mean(numeric_only=True))
pclass['Age'].agg(lambda x: x.isnull().sum())
````

### `Q-2:` Using `groupby` make groups using the `"Pclass"` column and fill every group's `"Embarked"` column's missing values with the mode value of that group. After that, print every group's `"Embarked"` column's value counts in ascending order.

````python
for group in pclass.groups.keys():
    pclass['Embarked'].get_group(group).fillna(pclass['Embarked'].get_group(group).mode())
    print(pclass['Embarked'].get_group(group).value_counts())
````

### `Q-4:` Perform `groupby` based on the `"Team"` column and then perform Z Normalization on top of the below columns of each group:
1. Passes
2. Passes Completed
3. Attempted Line Breaks
4. Completed Line Breaks

You have to make a python function named `z_normalization` which takes two arguments:

1. *group:* Every group that you have created
2. *cols_to_perform:* This parameter takes a list of columns on which you have to perform the Z-Normalization.

For this problem, you have to use th `apply()` method.

$\Large Z \ - \ Normalization = \frac{X_i - \mu}{std} $

After that find out the below values for each group:
- minimum "Passess"
- maximum "Passess"
- minimum "Yellow Cards"
- maximum "Yellow cards"
- average "Yellow Cards"
- maximum "Attempted Line Breaks"
- minimum "Attempted Line Breaks"
- standardard deviation of "Attempted Line Breaks"
- average Possession

````python
# Team_group = fifa_df.groupby('Team')
# cols_to_perform = ['Passes','Passes Completed','Attempted Line Breaks','Completed Line Breaks']

# def z_normalization(group,cols):
#     for col in cols:
#         std = group[col].std()
#         mean = group[col].mean()
#         group[f'{col}_z_norm'] = (group[col] - mean)/std
#         return group

# Team_group = Team_group.apply(z_normalization,cols_to_perform).groupby('Team')

# code here
def z_normalization(group, cols_to_perform):
    for col in cols_to_perform:
        std = group[col].std()
        mean = group[col].mean()
        group[f"{col}_z_norm"] = (group[col] - mean)/std
    return group


cols_to_perform = ["Passes", "Passes Completed", "Attempted Line Breaks", "Completed Line Breaks"]
groups = fifa_df.groupby("Team")
groups = groups.apply(z_normalization, cols_to_perform = cols_to_perform).groupby("Team")
groups.agg(
{
    "Passes": ["min", "max"],
    "Yellow Cards": ["min", "max", "mean"],
    "Attempted Line Breaks": ["max", "min", "std"],
    "Possession (%)": "mean"
})
````

### `Q-5:` Find batsman in below category-
* Highest score while chasing
* Best Strike rate while chasing and have faced 100+ balls


> Chasing mean team batting in second inning

````python
inning2 = ipl[ipl['innings'] == 2]
inning2.groupby(['batter','ID'])['batsman_run'].sum().sort_values(ascending=False).head(5)
````

````python
runs_group = inning2.groupby(['batter']).agg(
    {
        'batsman_run':'sum',
        'ballnumber':'count'
    }
)
runs_group['Strike rate'] = runs_group['batsman_run'] / runs_group['ballnumber'] * 100
runs_group[runs_group['ballnumber'] >= 100].sort_values('Strike rate',ascending=False)
````

````python
# Assistant
# Group the data by 'batter' and aggregate the total runs and count of balls faced
runs_group = inning2.groupby(['batter']).agg(
    {
        'batsman_run': 'sum',  # Sum of runs scored by each batter
        'ballnumber': 'count'   # Count of balls faced by each batter
    }
)

# Calculate the strike rate for each batter
runs_group['Strike rate'] = runs_group['batsman_run'] / runs_group['ballnumber'] * 100

# Filter the group to include only those batters who faced 100 or more balls and sort by strike rate in descending order
runs_group[runs_group['ballnumber'] >= 100].sort_values('Strike rate', ascending=False).head(5)
````

### `Q-8:` Make a dataframe for all batting pairs played together.
```
Batsman1 Batsman2 Runs Avg StrikeRate
```

> Just to ease this question you can count wide-balls for strike rate.

````python
ipl.head(4)
````

````python
pair = pd.DataFrame(columns=['Runs'])
pair['Runs'] = ipl.groupby(['batter','non-striker'])['batsman_run'].sum()
pair['Avg'] = ipl.groupby(['batter','non-striker'])['batsman_run'].mean()
pair['Strike rate'] = (ipl.groupby(['batter','non-striker'])['batsman_run'].sum() / ipl.groupby(['batter','non-striker'])['ballnumber'].sum()) * 100
pair.reset_index(inplace=True)
pair.rename(columns={
    'batter':'Batsman1',
    'non-striker':'Batsman2'
    }).sort_values(by='Runs',ascending=False)
````
